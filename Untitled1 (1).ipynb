Outlier detection and filtering

```python
from scipy import stats
import numpy as np

# Calculate Z-scores for a column, e.g., 'age'
merged_df['age_zscore'] = np.abs(stats.zscore(merged_df['age']))

# Filter out rows where the Z-score is greater than 3
filtered_df_z = merged_df[merged_df['age_zscore'] < 3]
print(filtered_df_z[['age', 'age_zscore']].head())
```

```python
# Calculate IQR for a column, e.g., 'age'
Q1 = merged_df['age'].quantile(0.25)
Q3 = merged_df['age'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
filtered_df_iqr = merged_df[(merged_df['age'] >= lower_bound) & (merged_df['age'] <= upper_bound)]
print(filtered_df_iqr[['age']].head())
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit the ARIMA model (p, d, q values need to be determined)
model = ARIMA(df['morekids'], order=(5, 1, 0))  # Adjust the order as needed
model_fit = model.fit()

# Make forecast (forecast for next 10 time steps)
forecast = model_fit.forecast(steps=10)
print(forecast)
```

```python
from sklearn.metrics import mean_squared_error

# Assuming we have actual values for comparison
actual = df['morekids'][-10:]  # Last 10 values
predicted = forecast

# Calculate RMSE
rmse = mean_squared_error(actual, predicted, squared=False)
print(f'Root Mean Squared Error (RMSE): {rmse}')
```

Optimal number of cluster using elbow method

```python
# Elbow Method to find optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(data_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow graph
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Applying K-Means with optimal clusters
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0)  # Adjust clusters based on elbow method
kmeans_labels = kmeans.fit_predict(data_scaled)

# Adding the cluster labels to the original data
df['KMeans_Cluster'] = kmeans_labels
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score

# Load your dataset
# Replace with your actual dataset path
# df = pd.read_csv('your_data.csv')  # Uncomment and replace with your actual file path

# Example DataFrame (replace with your actual data)
data = {
    'morekids': [1, 0, 1, 0, 1],  # Dependent variable (Y)
    'gender1': [1, 0, 0, 1, 0],
    'age': [23, 34, 45, 29, 36],
    'afam': [0, 1, 1, 0, 0],
    'hispanic': [0, 0, 1, 0, 1],
    'other': [1, 1, 0, 0, 1],
    'work': [1, 0, 1, 1, 0]
}

df = pd.DataFrame(data)

# Step 1: Feature selection (you can adjust these based on your data)
X = df[['gender1', 'age', 'afam', 'hispanic', 'other', 'work']]  # Independent variables
y = df['morekids']  # Dependent variable

# Step 2: Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Standardize the data (important for linear regression)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 4: Initialize the linear regression model
model = LinearRegression()

# Step 5: Train the model
model.fit(X_train_scaled, y_train)

# Step 6: Predict on the test set
y_pred = model.predict(X_test_scaled)

# Step 7: Model evaluation
# Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# R-squared (R)
r2 = r2_score(y_test, y_pred)

# Print the results
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')

# Step 8: Visualize the Actual vs Predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue', edgecolor='black')
plt.title('Actual vs Predicted Values')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()

# Step 9: Cross-validation (Optional)
cv_scores = cross_val_score(model, X, y, cv=5, scoring='neg_mean_squared_error')
print(f'Cross-validation MSE: {np.mean(np.abs(cv_scores))}')
```
