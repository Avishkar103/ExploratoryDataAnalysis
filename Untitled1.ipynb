Outlier detection and filtering

```python
from scipy import stats
import numpy as np

# Calculate Z-scores for a column, e.g., 'age'
merged_df['age_zscore'] = np.abs(stats.zscore(merged_df['age']))

# Filter out rows where the Z-score is greater than 3
filtered_df_z = merged_df[merged_df['age_zscore'] < 3]
print(filtered_df_z[['age', 'age_zscore']].head())
```

```python
# Calculate IQR for a column, e.g., 'age'
Q1 = merged_df['age'].quantile(0.25)
Q3 = merged_df['age'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
filtered_df_iqr = merged_df[(merged_df['age'] >= lower_bound) & (merged_df['age'] <= upper_bound)]
print(filtered_df_iqr[['age']].head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot a boxplot for 'age' to identify outliers
sns.boxplot(data=merged_df, x='age')
plt.show()
```

```python
from sklearn.ensemble import IsolationForest

# Use Isolation Forest on multiple columns, e.g., 'age' and 'work'
iso_forest = IsolationForest(contamination=0.05)  # Adjust contamination rate based on expected outliers
outliers = iso_forest.fit_predict(merged_df[['age', 'work']])

# Filter out detected outliers
merged_df['is_outlier'] = outliers
filtered_df_forest = merged_df[merged_df['is_outlier'] == 1]
print(filtered_df_forest.head())
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Fertility2 (1).csv')  # Adjust the path if necessary

# Create a sequential index (assuming row order represents time)
df['time_index'] = range(1, len(df) + 1)

# Set the 'time_index' as the index (this will be our time-based index)
df.set_index('time_index', inplace=True)

# Check the first few rows
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Plot a specific column over time (replace 'morekids' with your desired column)
plt.figure(figsize=(10, 6))
plt.plot(df['morekids'])  # You can replace 'morekids' with any other column
plt.xlabel('Time Index')
plt.ylabel('More Kids')
plt.title('Time Series Plot of More Kids')
plt.show()
```

```python
# Convert 'morekids' column (if it contains categorical values like 'yes'/'no') to numeric
df['morekids'] = df['morekids'].map({'yes': 1, 'no': 0})

# Check the first few rows to confirm the conversion
print(df[['morekids']].head())
```

```python
# Drop rows with missing values in 'morekids' or fill them with the median
df['morekids'] = df['morekids'].fillna(df['morekids'].median())

```

```python
# Decompose the time series (additive model)
decomposition = sm.tsa.seasonal_decompose(df['morekids'], model='additive', period=12)  # Adjust period if needed
decomposition.plot()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform Augmented Dickey-Fuller Test to check stationarity
result = adfuller(df['morekids'].dropna())  # Drop any NaN values

print('ADF Statistic:', result[0])
print('p-value:', result[1])

# If the p-value is less than 0.05, the series is stationary.
```

```python
from pandas.plotting import autocorrelation_plot

# Plot autocorrelation
autocorrelation_plot(df['morekids'])
plt.show()
```

```python
# Calculate rolling mean (window size 7)
rolling_mean = df['morekids'].rolling(window=7).mean()

# Plot original series and rolling mean
plt.figure(figsize=(10, 6))
plt.plot(df['morekids'], label='Original')
plt.plot(rolling_mean, label='Rolling Mean', linewidth=3)
plt.legend()
plt.title('Rolling Mean of More Kids')
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit the ARIMA model (p, d, q values need to be determined)
model = ARIMA(df['morekids'], order=(5, 1, 0))  # Adjust the order as needed
model_fit = model.fit()

# Make forecast (forecast for next 10 time steps)
forecast = model_fit.forecast(steps=10)
print(forecast)
```

```python
from sklearn.metrics import mean_squared_error

# Assuming we have actual values for comparison
actual = df['morekids'][-10:]  # Last 10 values
predicted = forecast

# Calculate RMSE
rmse = mean_squared_error(actual, predicted, squared=False)
print(f'Root Mean Squared Error (RMSE): {rmse}')
```

```python
# Save forecasted results to a CSV
forecast_df = pd.DataFrame({'time_index': range(df.index[-1] + 1, df.index[-1] + 11), 'forecast': forecast})
forecast_df.to_csv('forecast.csv', index=False)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Fertility2 (1).csv')  # Adjust the path if necessary

# Convert 'morekids' column from categorical to numeric (if necessary)
df['morekids'] = df['morekids'].map({'yes': 1, 'no': 0})

# Create a synthetic time index (assuming rows represent time)
df['time_index'] = range(1, len(df) + 1)

# Set the 'time_index' as the index for time-based indexing
df.set_index('time_index', inplace=True)

# Check the first few rows
print(df.head())

```

```python
import matplotlib.pyplot as plt

# Plot a specific column (e.g., 'morekids') over the time index
plt.figure(figsize=(10, 6))
plt.plot(df['morekids'], label='More Kids')
plt.xlabel('Time Index')
plt.ylabel('More Kids')
plt.title('Time Series of More Kids')
plt.legend()
plt.show()
```

```python
# Grouping by a synthetic time period, for example, every 10 time points
df['time_period'] = (df.index - 1) // 10  # Group every 10 rows (adjust as needed)

# Group by the 'time_period' and calculate the mean for each group
grouped_data = df.groupby('time_period')['morekids'].mean()

# Plot the grouped data
plt.figure(figsize=(10, 6))
plt.plot(grouped_data, label='Grouped Mean of More Kids')
plt.xlabel('Time Period')
plt.ylabel('Average More Kids')
plt.title('Grouped Time Series Data')
plt.legend()
plt.show()
```

```python
# Create a datetime index (assuming each row represents a day)
df['datetime'] = pd.date_range(start='2020-01-01', periods=len(df), freq='D')
df.set_index('datetime', inplace=True)

# Now resample the data to monthly frequency (e.g., resampling by month)
resampled_df = df['morekids'].resample('M').mean()  # Mean value for each month

# Plot the resampled data
plt.figure(figsize=(10, 6))
resampled_df.plot(label='Monthly Resampled Data', color='orange')
plt.xlabel('Date')
plt.ylabel('Average More Kids')
plt.title('Monthly Resampling of More Kids')
plt.legend()
plt.show()
```
